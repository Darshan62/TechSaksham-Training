```python
#Top 10 Highest Run Scorers in IPL
top_scorers = df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_scorers.index, y=top_scorers.values,palette='magma')
plt.xlabel("Batsmen")
plt.ylabel("Total Runs")
plt.title("Top 10 Highest Run Scorers in IPL")
plt.xticks(rotation=45)
plt.show()
```

```python
#Most Common Dismissals in IPL
dismissal_types = df['dismissal_kind'].value_counts()

plt.figure(figsize=(8,4))
sns.barplot(x=dismissal_types.index, y=dismissal_types.values, palette='plasma')
plt.xlabel("Dismissal Type")
plt.ylabel("Count")
plt.title("Most Common Dismissals in IPL")
plt.xticks(rotation=45)
plt.show()
```

```python
#Runs Scored in Each Over (Powerplay vs Death Overs)
over_wise_runs = df.groupby('over')['total_runs'].sum()

plt.figure(figsize=(10,5))
sns.lineplot(x=over_wise_runs.index, y=over_wise_runs.values, marker='o', color='blue')
plt.xlabel("Over Number")
plt.ylabel("Total Runs")
plt.title("Runs Scored in Each Over (Powerplay vs Death Overs)")
plt.xticks(range(1, 21))
plt.grid(True)
plt.show()
```

```python

# Scatter Plot of Runs Scored vs. Overs
plt.figure(figsize=(10, 5))
plt.scatter(df['over'], df['batsman_runs'], alpha=0.5, color='blue')

# Labels and Title
plt.xlabel("Over", fontsize=12)
plt.ylabel("Runs Scored", fontsize=12)
plt.title("Scatter Plot of Runs Scored vs. Overs", fontsize=14)
plt.show()
```

```python


# Function to create a bar plot for categorical columns
def bar_plot(variable):
    var = df[variable]
    varValue = var.value_counts()  # Count occurrences of each category
    plt.figure(figsize=(10, 5))
    plt.bar(varValue.index, varValue, color='skyblue')  # Bar plot
    plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability
    plt.ylabel("Frequency", fontsize=12)
    plt.xlabel(variable, fontsize=12)
    plt.title(f"Bar Plot of {variable}", fontsize=14)
    plt.show()
    print(f"{variable}: \n{varValue}")

# List of categorical columns to visualize
columns_plot = ['batting_team', 'bowling_team', 'dismissal_kind', 'extras_type']

# Generate bar plots for each selected column
for i in columns_plot:
    bar_plot(i)
```

```python


# Function to create a histogram for numerical columns
def hist_plot(variable):
    plt.figure(figsize=(10, 5))
    plt.hist(df[variable], bins=20, color='orange', edgecolor='black')  # Histogram with bins
    plt.xlabel(variable, fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.title(f"Histogram of {variable}", fontsize=14)
    plt.show()
    print(f"{variable}:\n{df[variable].describe()}")  # Print summary statistics

# List of numerical columns to visualize
columns_plot = ['batsman_runs', 'total_runs', 'extra_runs', 'over', 'ball']

# Generate histogram plots for each selected column
for i in columns_plot:
    hist_plot(i)
```

```python
#Using .fillna() to Fill Missing Values
df['extras_type'].fillna("No Extra", inplace=True)
df['player_dismissed'].fillna("Not Out", inplace=True)
df['dismissal_kind'].fillna("Not Out", inplace=True)
df['fielder'].fillna("No Fielder Involved", inplace=True)
```

```python
print(df.isna().sum())
```

```python
to_drop = ['extras_type', 'fielder']

# Drop only if columns exist
df = df.drop(columns=[col for col in to_drop if col in df.columns], axis=1)
```

```python
df = pd.get_dummies(df, columns=['batting_team', 'bowling_team', 'dismissal_kind'], drop_first=True)
```

```python
# Selecting only numerical columns
numeric_df = df.select_dtypes(include=['number'])

# Creating the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

for col in ['batter', 'bowler', 'non_striker', 'player_dismissed']:
    df[col] = LabelEncoder().fit_transform(df[col])
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
# Counting wins when a team batted first (Inning 1)
batting_first_wins = df[df['inning'] == 1].groupby('batting_team')['match_id'].nunique()

# Convert to DataFrame
batting_first_wins = batting_first_wins.reset_index()
batting_first_wins.columns = ['Team', 'Wins']

# Plot the data
plt.figure(figsize=(12,6))
sns.barplot(x='Team', y='Wins', data=batting_first_wins, palette='Reds_r')

# Labels and title
plt.xlabel("Teams")
plt.ylabel("Total Wins")
plt.title("Total Wins When Batting First")
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtering matches where team batted first and won
bat_first_wins = df[df['inning'] == 1].groupby('batting_team')['match_id'].nunique().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=bat_first_wins.index, y=bat_first_wins.values, palette='Blues_r')

# Labels
plt.xlabel("Teams")
plt.ylabel("Wins when Batting First")
plt.title("Number of Matches Won by Teams When Batting First")
plt.xticks(rotation=45)
plt.show()
```

```python
# Grouping by over to count wickets
wickets_per_over = df[df['is_wicket'] == 1].groupby('over')['is_wicket'].count()

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=wickets_per_over.index, y=wickets_per_over.values, palette='cividis')

# Labels
plt.xlabel("Over Number")
plt.ylabel("Total Wickets Taken")
plt.title("Most Wicket-Taking Overs in IPL")
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by bowling team and sum up extra runs conceded
extra_runs_by_team = df.groupby('bowling_team')['extra_runs'].sum().sort_values(ascending=False)

# Plot the extra runs conceded by teams
plt.figure(figsize=(12, 6))
sns.barplot(x=extra_runs_by_team.index, y=extra_runs_by_team.values, palette="coolwarm")

# Add labels
plt.xlabel("Teams", fontsize=12)
plt.ylabel("Total Extra Runs Conceded", fontsize=12)
plt.title("Most Extra Runs Conceded by Teams in IPL", fontsize=14)
plt.xticks(rotation=45, ha="right")

# Show plot
plt.show()
```
